```python
from vipy.image import ImageDetection
from vipy.image import Scene
from vipy.object import Detection
```

```python
url='https://upload.wikimedia.org/wikipedia/commons/thumb/2/23/Bubo_virginianus_06.jpg/1920px-Bubo_virginianus_06.jpg'
im = ImageDetection(url=url, category='Great Horned Owl').boundingbox(xmin=300, ymin=320, width=1200, height=1600)
```

```python
help(im.resize)
im.resize(cols=512).show()
```

```python
help(im.centersquare)
im.centersquare().show()
```

```python
help(im.dilate)
im.dilate(0.75).show()
```

```python
help(im.crop)
help(im.maxdim)
im.crop().maxdim(256).show()
```

```python
help(im.zeropad)
im.zeropad(64,64).show()
```

```python
im = Scene(url=url, category='Great Horned Owl', objects=[Detection('left eye', xmin=600, ymin=800, width=250, height=250), 
                                                          Detection('right eye', xmin=1000, ymin=800, width=250, height=250)])
print(im)
print(im.objects())
```

```python
im.mindim(512).show()
```

```python
help(im.fliplr)
im.fliplr().show()
```
