```python
import vipy.video
from vipy.object import Track
from vipy.geometry import BoundingBox
```

```python
bb = BoundingBox(xmin=280,ymin=850,width=600,height=300)
track = Track(category='Person', keyframes=[100], boxes=[bb])
print(track)
```

```python
mp4file = '../test/Video.mp4'
vid = vipy.video.Scene(filename=mp4file, tracks=[track])
print(vid)
```

```python
vid.load()
```
