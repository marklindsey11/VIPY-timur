```python
from vipy.dataset.kinetics import Kinetics700
from vipy.batch import Batch
```

```python
kinetics = Kinetics700('/tmp/kinetics').download(verbose=False)
```

```python
trainset = kinetics.trainset()
print(trainset[0])
print(trainset[1])
```

```python
trainbatch = Batch(trainset[0:2], n_processes=2).map(lambda v: v.download(ignoreErrors=True)).batch()
```

```python
print(trainbatch[0])
video = trainbatch[0].resize(cols=512).load()
frame = video[150]
frame.show()
```
