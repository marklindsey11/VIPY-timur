```python
import vipy
```

```python
# Load an image from wikipedia and create three objects
# Transform it in three ways and show the image
im = vipy.image.owl().mindim(512).show(figure=1).fliplr().show(figure=2).minsquare().show(figure=3)
```

```python
# The images are shown inline in a notebook, but from ipython you can close all windows using:
im.closeall()
```
